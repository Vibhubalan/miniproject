**USING ROBERTA MODEL**

```python
from transformers import AutoTokenizer
from scipy.special import softmax
from torch import nn
from transformers import AutoModelForSequenceClassification
```



*   Transmformer library- NLP models which consists models like Roberta,bert,
*   Importing AutoTokenizer and AutoModelForSequenceClassification from transformers library
* AutoTokenizer-Used to tokenize the input text data
* AutoModelForSequenceClassification- Used for loading pretrained models
* From the Scipy.special where scipy is a computing library and here special contains multiple function for scientific computation
* The softmax function is a mathematical function that takes as input a vector of numerical scores and outputs another vector of the same length.




```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
#VADER results which we declared as "eg"
print(eg)
sia.polarity_scores(eg)
```

```python
#now running the same result in 'eg' using roberta model
tokenizer(eg,return_tensors='pt')
#we get the encode text as result and only models can understand
```

```python
encoded_text=tokenizer(eg,return_tensors='pt')
output=model(**encoded_text)
output
#we get the output as tensor
```

```python
#now from tensor to numpy
encoded_text=tokenizer(eg,return_tensors='pt')
output=model(**encoded_text)
scores=output[0][0].detach().numpy()
scores=softmax(scores)
scores
#we get output as numpy array
```

```python
#gives the analyis from roberta model
encoded_text=tokenizer(eg,return_tensors='pt')
output=model(**encoded_text)
scores=output[0][0].detach().numpy()
scores=softmax(scores)
scores_dict={
    'roberta_neg':scores[0],
    'roberta_neu':scores[1],
    'roberta_pos':scores[2]
}
print(scores_dict)
#we can see that roberta model gives much more than vader model showing roberta>vader
```

```python
#running on the entire dataset with roberta model
def polarity_scores_roberta(eg):
  encoded_text=tokenizer(eg,return_tensors='pt')
  output=model(**encoded_text)
  scores=output[0][0].detach().numpy()
  scores=softmax(scores)
  scores_dict={
    'roberta_neg':scores[0],
    'roberta_neu':scores[1],
    'roberta_pos':scores[2]
}
  return scores_dict
```

```python
res={}
for i,row in tqdm(df.iterrows(),total=len(df)):
    review=row['review']
    name=row['name']
    vader_result=sia.polarity_scores(review)
    vader_result_rename = {}
    for key, value in vader_result.items():
        vader_result_rename[f"vader_{key}"] = value
    roberta_result=polarity_scores_roberta(review)
    both={**vader_result_rename,**roberta_result}
    res[name]=both
```

```python

results_df=pd.DataFrame(res).T
results_df=results_df.reset_index().rename(columns={'index':'name'})
results_df=results_df.merge(df,how='left')
```

```python

# Here we are defining the variables to plot
vars_to_plot = ['vader_neg', 'vader_neu', 'vader_pos', 'roberta_neg', 'roberta_neu', 'roberta_pos']
num_rows = 2
num_cols = 3
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 10))
axs = axs.flatten()

# Creating bar plots for each variable
for i, var in enumerate(vars_to_plot):
    sns.barplot(data=results_df, x='rating', y=var, hue='rating', palette='tab10', ci=None, ax=axs[i])
    axs[i].set_xlabel('Rating')
    axs[i].set_ylabel(var)
    axs[i].set_title(f'Average {var} by Rating')
    axs[i].legend([], frameon=False)

plt.tight_layout()
plt.show()
```
